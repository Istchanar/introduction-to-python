```python
from sympy import *
from sympy.plotting import plot
init_printing()

# function
x = Symbol('x')
func = 0.6*x**3 + 5.5*x**2 + 10*x - 5
```

```python
roots = solveset(func, x)
print('roots:', roots)

critical_points = solveset(func.diff(x), x)
print('critical points:', critical_points)

interval_first = is_increasing(func, Interval.open(-oo, -5.1))
interval_second = is_decreasing(func, Interval.open(-4.9, -1.2))
interval_third = is_increasing(func, Interval.open(-1.1, +oo))
print(f'dic/increasing check: dec (-5.0, -1.11): {interval_first}; inc: (-oo, -5.0) and (-1.11, +oo): {interval_second, interval_third}')

critical_points_y = [func.subs(x, a) for a in critical_points]
print('vertexes on Y:', critical_points_y)

func_positive = {'(-6.2, -3.25)': func.subs(x, -5), '(0.4, +oo)': func.subs(x, 1)}
func_negative = {'(-oo, -6.2)': func.subs(x, -7), '(-3.25, 0.4)': func.subs(x, -1)}
print('positive func values:', func_positive)
print('negative func values:', func_negative)
```
